# Notebook de los datos de turismo mundial por Pais y año  desde 1995 a 2023

```python
# Importo las librerías necesarias
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
# Accedo a la url de datosmacro, para recolectar los nombres de los paises para construir la url de cada pais para acceder a sus datos. 
url = 'https://datosmacro.expansion.com/comercio/turismo-internacional'
respuesta = requests.get(url)
respuesta.status_code
```

```python
html = respuesta.content
soup = bs(html, 'lxml')
```

```python
# Accedo la tabla donde están los paises, y guardo cada uno de sus tag 'a', para recorerlos despues y quedarme con todos los nombres
lista_paises = []
paises =  soup.find('table', id="tb1")
paises = paises.find_all('tr')
for pais in paises:
    for a in pais.find_all('a'):
        lista_paises.append(a.text[:-4].lower())

```

```python
# Recorro cada ta 'a', quedandome con el nombre del país y modificandolos para cambiar los espacios por '-' y quitando las 'y'
lista_paises_new = []
for pais in lista_paises:
    pais = pais.split()
    if 'y' in pais:
        pais.remove('y')
    pais = '-'.join(pais)
    lista_paises_new.append(pais)
```

### Una vez conseguidos todos los nombres de paises preparados para construir las urls, recorro cada uno de ellos, accediendo a su código html.
### Me quedo con la tabla que contiene los datos que quiero y genero un diccionario con las claves los años y los valores los totales de turistas de ese año.

```python
lista_paises = []
lista_paises_sin_datos = []
for pais in lista_paises_new:
    try:
        url = f'https://datosmacro.expansion.com/comercio/turismo-internacional/{pais}'
        respuesta = requests.get(url)
        html_pais = respuesta.content
        soup_pais = bs(html_pais, 'lxml') 
        datos =  soup_pais.find('table', id="tb0")#driver.find_elements(By.XPATH, '//*[@id="tb1"]/tbody/tr/td/a')
        datos = datos.find_all('tr')
        num_pais = {x: None for x in range(1995,2025)}
        for dato in datos[1:]:
            num_pais[int(dato.text[:4])] = int(dato.text[4:].replace('.', ''))
        num_pais['pais'] = pais
        lista_paises.append(num_pais)
        print(f'{pais} scrapeado con exito')
    except:
        print(f'El pais {pais} no pudo ser scrapeado')
        lista_paises_sin_datos.append(pais)
        
```

```python
# Una vez que tengo todas las páginas y todos los datos que necesitaba creo un dataframe con la lista.
df = pd.DataFrame(lista_paises)
df = df.set_index('pais').reset_index()
```
